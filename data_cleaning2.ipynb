```python
missing_value = ["N/a",'na',np.nan]
df = pd.read_csv('data-jumlah-sarana-angkut-sampah-dan-tinja-air-kotor-provinsi-dki-jakarta-tahun-2011-2014.csv',na_values=missing_value)
```

```python
df.isnull().any()
```

```python
sns.heatmap(df.isnull(), yticklabels=False,cbar=False,cmap='viridis') #cmap=['magma','inferno','plasma']
```

```python
sns.heatmap(df.isnull(), yticklabels=False)
```

```python
#menghapus data yang NaN 
df.dropna()
```

```python
df.fillna(0)
```

```python
#cara 1 untuk mengganti kolom jumlah yang kosong dengan jumlah rata-rata
df['jumlah'].fillna(value=df['jumlah'].mean(), inplace=True)
sns.heatmap(df.isnull(), yticklabels=False, cbar=False, cmap='viridis')
df
```

```python
#cara 2 untuk mengganti kolom jumlah yang kosong dengan unknown
df.jumlah = df.jumlah.fillna('unknown')
sns.heatmap(df.isnull(), yticklabels=False, cbar=False, cmap='viridis')
df
```

```python
#cara 3 untuk menghapus kolom jumlah yang kosong
df.drop('jumlah',axis=1,inplace=True)
sns.heatmap(df.isnull(), yticklabels=False, cbar=False, cmap='viridis')
df
```
